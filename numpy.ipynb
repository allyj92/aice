## 비교와 브로드캐스팅

- 브로드캐스팅 순서
   - 1. 차원 맞추기 : (2,3)+(3,) =>(2,3)+(1,3)
     2. 모양 맞추기 : (2,3)+(1,3) => (2,3)+(2,3)

```python
np.array([1,2,3,4,5]) == 1
```

```python
np.array([1,2,3,4,5]) == np.array([1,2,3,4,6])
```

```python
np.array([[1,2,3,4,5],[10,20,30,40,50]]) == np.array([1,2,3,4,5])
```

```python
np.array([[1,2,3,4,5],[10,20,30,40,50]]) + np.array([1,2,3,4,6])
```

```python
np.array([[1,2,3,4,5],[10,20,30,40,50]]) + np.array([1,2]).reshape(-1,1)
```

```python
np.array([1,2,3,4,5]) + np.array([1,2,3,4,5]).reshape(-1,1)
```
